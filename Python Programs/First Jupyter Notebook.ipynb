```python
#String Manipulation
string = "i aM gauRav aGarwal"
string2 = "Hi, "
print ("original string           = ", string)
print ("in lowercase              = ", string.lower())
print ("IN UPPERCASE              = ", string.upper())
print ("In Title Case             = ", string.title())
print ("number of characters      = ", len(string), "(including spaces)")
print ("number of A's             = ", string.count("a"))
print ("Concatenation             = ", string2 + string.title())

split_string = string.split(" ")
print ("Split                     = ", split_string)
print ("Join with '@'             = ", "@".join(split_string))

#Slicing
print ("First 7 character         = ", string[0:7], "(including spaces)")
print ("Reverse Slcing            = ", string[::-1])

#F - strings
print (f"the string used are  =  {string} and {string2}")
```
