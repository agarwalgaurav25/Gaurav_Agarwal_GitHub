```python
#Program to print each line of file in reverse order

string = """
Good Afternoon
My name is Alex.
I a 7 years old."""
print (string[::-1])
```

```python
#Program to use to use and join methods in the string and trace a birthday with a dictionary

date = str(input("Enter Birthday in dd/mm/yy format :"))
birthday = {date:""}
L1 = date.split("/")
new_date = "-".join(L1)
birthday[date]= new_date
print(birthday)
```

```python
#Program to use Split and Join in the string

string = "I am Gaurav Agarwal"
split_string = string.split(" ")
print ("Original String = ", string)
print ("Split           = ", split_string)
print ("Join with ' @ ' = ", " @ ".join(split_string))
```

```python
#Program to count the number of lower and uppercase letters

string = "Good Afternoon! Today is Thursday"
upper = 0
lower = 0
for i in string:
    if i.isupper():
        upper += 1
    elif i.islower():
        lower +=1
print ("Uppercase = ", upper)
print ("Lowercase = ", lower)
```

```python
#Program to count the number of characters in a word and map
#it to a dictionary

string ="I love India"
words = string.split()
word_count = {}
for word in words:
    word_count[word] = len(word)
print(word_count)
```

**FUNCTIONS QUESTIONS**

```python
def fibonacci_sequence(n):
  a, b = 0, 1
  count = 0
  fib_list = []
  while count < n:
    fib_list.append(a)
    nth = a + b
    a = b
    b = nth
    count += 1
  print(fib_list)
n = int(input("enter the number of elements needed : "))
fibonacci_sequence(n)
```

```python
#Function to identify number as odd or even
def check(num):
  if num == 0 :
    return "0 is neither even nor odd"
  elif num % 2 == 0:
    return "number is even"
  else:
    return "number is odd"

num = int(input("enter the number : "))
print(check(num))
```

```python
#Function to find the factorial of a number
def factorial(num):
  fact = 1
  i = 1
  while i <= num:
    fact = fact * i
    i += 1
  print(f"Factorial of {num} is {fact}")

num = int(input("Enter the number to get the factorial for : "))
factorial(num)
```

```python
#Function to multiply all the numbers in a list
L1 = [1,4,5,-2,3,8]
prod = 1
for i in L1:
  prod = prod * i
print(f"The product of {L1} is {prod}")
```

```python
#Function to find the sum of all the number
L1 = [1,4,5,-2,3,8]
snum = 0
for i in L1:
  snum += i
print(f"The Sum of {L1} is {snum}")
```
